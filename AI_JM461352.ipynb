# Metoda 1 - regresja logistyczna

Model regresji logistycznej to klasyfikator, który przewiduje prawdopodobieństwo przynależności do jednej z dwóch klas. W tym przypadku celem jest przewidzenie, czy pasażer Titanica przeżył (1) czy nie (0). Model używa funkcji logistycznej (sigmoid), która zwraca wartości między 0 a 1, reprezentujące prawdopodobieństwo. Na podstawie tego prawdopodobieństwa dokonuje klasyfikacji. 

Metoda cechuje się zaletami w postaci dobrej efektywności, łatwości interpretacji, dobrej stabilności. Jest natomiast wrażliwa na złożone zależności miedzy zminnymi a wynikiem, oraz ogranicza ją klasyfikacja binarna, natomiast jest dobrze dopasowana do prezentowanego przypadku, gdzie atrybut doecelowy jest binarny.


# Schemat

- Dane wejściowe:  
$x_1, x_2, \dots, x_n \$ → Cechy (wiek, płeć, klasa itd.)

- Model:

$ z = \beta_0 + \beta_1 x_1 + \dots + \beta_n x_n $

- Sigmoid (logistyczna):

$P(y=1) = \frac{1}{1 + e^{-z}} $

- Decyzja: 
    - Jeśli $P(y=1) > 0.5 $, przewidujemy 1 (przeżył).
    - Jeśli $ P(y=1) \leq 0.5 $, przewidujemy 0 (nie przeżył).


# Zastosowane metryki


### Dokładność (Accuracy):
Dokładność to podstawowa miara oceny klasyfikatora, która wyraża stosunek liczby prawidłowych przewidywań do całkowitej liczby próbek. Można ją opisać wzorem:


${Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$



Gdzie:
- **TP** (True Positives) – liczba prawdziwie pozytywnych klasyfikacji,
- **TN** (True Negatives) – liczba prawdziwie negatywnych klasyfikacji,
- **FP** (False Positives) – liczba fałszywie pozytywnych klasyfikacji,
- **FN** (False Negatives) – liczba fałszywie negatywnych klasyfikacji.

Należy jednak pamiętać, że w przypadku niezbalansowanych danych (gdy liczba próbek jednej klasy znacznie przewyższa liczbę próbek innej klasy), accuracy może być mylące. Model może przewidywać dominującą klasę z wysoką dokładnością, ignorując mniejszą klasę, co prowadzi do słabych jakościowo i pozornie dobrych wyników. Dlatego warto korzystać z dodatkowych metryk.



### Precyzja (Precision):
Precyzja jest miarą tego, jak dokładny jest model w przewidywaniu klasy pozytywnej. Wyraża ona stosunek liczby prawdziwie pozytywnych przewidywań do liczby wszystkich przewidywań, które model oznaczył jako pozytywne. Można ją opisać wzorem:


${Precision} = \frac{TP}{TP + FP}$


Wysoka precyzja oznacza, że model rzadko popełnia błędy w klasyfikacji negatywnych próbek jako pozytywnych. Jest to kluczowe w zadaniach, w których fałszywie pozytywne wyniki są kosztowne.

### Recall (Czułość):
Recall to miara zdolności modelu do prawidłowego wykrycia wszystkich próbek należących do klasy pozytywnej. Wyraża stosunek liczby prawdziwie pozytywnych klasyfikacji do liczby wszystkich rzeczywistych pozytywnych próbek (zarówno prawdziwie, jak i fałszywie negatywnych). Można ją opisać wzorem:


$\text{Recall} = \frac{TP}{TP + FN}$


Wysoki recall oznacza, że model jest w stanie zidentyfikować większość pozytywnych przypadków, co jest istotne w zadaniach, w których fałszywie negatywne wyniki są niepożądane.


### F1-Score:
F1-Score to harmoniczna średnia precyzji i recall. Jest szczególnie użyteczna w sytuacjach, gdy mamy do czynienia z nierównomiernie zbalansowanymi danymi i zależy nam zarówno na wysokiej precyzji, jak i recall. F1-Score można opisać wzorem:


${F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$

F1-Score osiąga wartość 1, gdy zarówno precyzja, jak i recall są wysokie, a zbliża się do 0, gdy którykolwiek z tych wskaźników jest niski.


Podsumowując:
- **Accuracy** jest łatwa do zrozumienia, ale może być myląca w przypadku niezbalansowanych danych.
- **Precision** i **Recall** pozwalają lepiej zrozumieć wydajność modelu w przypadku klasyfikacji pozytywnych próbek, zwłaszcza w sytuacjach, gdzie fałszywe przewidywania (FP lub FN) są kosztowne.
- **F1-Score** daje dobrą miarę równowagi między precyzją a recall, co jest istotne w przypadku nierównomiernych danych.

# Metoda 2 - Naive Bayes

Naive Bayes to metoda klasyfikacji oparta na twierdzeniu Bayesa, zakładająca, że cechy są niezależne od siebie, co jest założeniem "naive" (naiwnym). Model ocenia prawdopodobieństwo przynależności do danej klasy na podstawie warunkowego prawdopodobieństwa cech przy założeniu ich niezależności. Jest szeroko stosowany w klasyfikacji tekstu, filtracji spamu oraz wielu innych problemach klasyfikacyjnych.
Metoda charakteryzuje się prostotą, szybkością i efektywnością przy dużych zbiorach danych. W przypadku danych z różnymi rodzajami cech (ciągłe, binarne, kategoryczne), Naive Bayes może być stosowany z różnymi wariantami obliczeń prawdopodobieństw. Jednak jest wrażliwy na nieprawidłowe założenia o niezależności cech, co może wpłynąć na dokładność klasyfikacji.

Implementacja metody została wykonana na podstawie wykładów, oraz "Sztuczna inteligencja dla inżynierów. Metody ogólne"  Mieczysław Muraszkiewicz (red.), Robert Nowak (red.)

# Schemat

- Dane wejściowe:
$x_1, x_2, \dots, x_n$ → Cechy (ciągłe, binarne, kategoryczne)

- Prawdopodobieństwo klasy:

$P(C_k) = \frac{N_{C_k}}{N}$

    Gdzie 
 $P(C_k)$ 
 to prawdopodobieństwo klasy $C_k$, $N_{C_k}$ to liczba próbek w klasie $C_k$, a $N$ to całkowita liczba próbek.

- Prawdopodobieństwo cech przy danej klasie:

  - Gaussian Naive Bayes (cechy ciągłe):

$P(x_i | C_k) = \frac{1}{\sqrt{2 \pi \sigma_{i}^2}} \exp \left( - \frac{(x_i - \mu_{i})^2}{2 \sigma_{i}^2} \right)$

Gdzie $\mu_{i}$ i $\sigma_{i}^2$ to średnia i wariancja cechy $x_i$ w klasie $C_k$.

 - Bernoulli Naive Bayes (cechy binarne):

$P(x_i | C_k) = p_{i} ^{x_i} (1 - p_{i})^{(1 - x_i)}$

Gdzie $p_{i}$ to prawdopodobieństwo cechy $x_i$ w klasie $C_k$.

 - Categorical Naive Bayes (cechy kategoryczne):

$P(x_i | C_k) = P(x_i = v_j | C_k)$

Gdzie $P(x_i = v_j | C_k)$ to prawdopodobieństwo, że cecha $x_i$ przyjmuje wartość $v_j$ w klasie $C_k$.

- Ostateczne prawdopodobieństwo klasy:

$P(C_k | x_1, x_2, \dots, x_n) \propto P(C_k) \prod_{i=1}^n P(x_i | C_k)$

 - Klasyfikacja polega na wyborze klasy, która maksymalizuje to prawdopodobieństwo.

- Decyzja:

 - Przewidujemy klasę $C_{k*}$, która maksymalizuje prawdopodobieństwo posteriori:

$C_{k*} = \arg \max_{C_k} \left[ P(C_k) \prod_{i=1}^n P(x_i | C_k) \right]$
